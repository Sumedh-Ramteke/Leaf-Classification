```python
from features import *
import pandas as pd
import os
```

```python
def generate_fetaures(dataset_path) -> pd.DataFrame:
    all_features = []
    for class_folder in os.listdir(dataset_path):
        class_path = os.path.join(dataset_path, class_folder)
        if not os.path.isdir(class_path):
            continue

        for image_name in os.listdir(class_path):
            image_path = os.path.join(class_path, image_name)
            features = extract_features(image_path)
            if features:
                features['Image'] = image_name
                features['Class'] = class_folder  # Label with class name
                all_features.append(features)
            print(f'{image_name} in {class_folder} processed...')
    df = pd.DataFrame(all_features)
    return df
```

# Generating Features for Train set

```python
train_path = "dataset/train"
```

```python
train_df = generate_fetaures(train_path)
train_df.head()
```

```python
train_df.to_csv("generated_features_train.csv", index=False)
print(f"Saved 'train_df' to 'generated_features_train'")
```

# Generating Features for Validation set

```python
valid_path = "dataset/valid"
```

```python
valid_df = generate_fetaures(valid_path)
valid_df.head()
```
